```python
import csv
```

```python
dictionary = {}
starting_year = 1880
end_year = 2021
for i in range(starting_year, end_year+1):
    print(i)
    with open(f'gender_census_data/yob{i}.txt', newline='') as file:
        reader = csv.reader(file)
        for row in reader:
            name = row[0].lower()
            if name not in dictionary.keys():
                dictionary.update({name: {'F': 0, 'M': 0}})
            curr = dictionary[name][row[1]]
            dictionary[name][row[1]] = curr + int(row[2])
```

```python
dictionary
```

```python
male_pct = []
for k, v in dictionary.items():
    male_pct.append({'first_name': k, 'male_pct': v.get('M')/(v.get('F') + v.get('M')) * 100})
```

```python
male_pct
```

```python
with open('census_names_w_pct.csv', 'w', newline='') as csvfile:
    fieldnames = ['first_name', 'male_pct']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for item in male_pct:
        writer.writerow(item)
```
