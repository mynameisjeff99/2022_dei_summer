```python
from scraping import Scraping
import sys
import json
```

# Test

```python
scraping = Scraping()
```

```python
with open(sys.path[0] + '/../data/columbia_directories_sans_med.json') as f:
    columbia = json.load(f)
```

```python
for department in columbia:
    url = department.get('faculty_page')
    try:
        res = scraping.get_department_info(url)
        department.update({'success': True, 'profiles': res})
    except:
        department.update({'success': False})
```

```python
with open(sys.path[0] + '/../data/columbia_test_sans_processing.json', 'w') as f:
    json.dump(columbia, f)
```

# Result

```python
completely_failed = []
for department in columbia:
    if department.get('success') == False:
        completely_failed.append(department.get('faculty_page'))
    else:
        res = department.get('profiles')
        print("____________________________________________")
        print(f"url: {department.get('faculty_page')}")
        print(f"len: {len(res)}")
        print(res[1])
        print(res[-1])
```

```python
completely_failed
```

```python
also_failed = ["https://www.biology.columbia.edu/people/faculty-department",
               "https://www.ee.columbia.edu/people-by-type/faculty"
              ]
```

```python
failed_directories = []
for department in columbia:
    url = department.get('faculty_page')
    if url in completely_failed or url in also_failed:
        failed_directories.append(department)
```

```python
with open(sys.path[0] + '/../data/failed_directories.json', 'w') as f:
    json.dump(failed_directories, f)
```

```python
failed_rate = len(failed_directories) / len(columbia)
print(f"success rate for testing ({len(columbia)} departments)= {1 - failed_rate:.2f}")
```
