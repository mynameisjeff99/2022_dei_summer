```python
import sys
import json
```

```python
schools = ["columbia", "dartmouth", "yale", "princeton", "brown", "cornell", "upenn", "harvard"]
```

```python
def dedupe(school):
    for department in school:
        if department.get('success') == True:
            ind = 0
            names = []
            profiles = department.get('profiles')
            print(len(profiles))
            while True:
                try:
                    prof = profiles[ind]
                    the_name = prof.get('name')
                    if the_name in names:
                        department.get('profiles').pop(ind)
                    else:
                        names.append(the_name)
                        ind += 1
                except:
                    break
            if len(profiles) == 1:
                department.update({'success': False, 'profiles': None})
```

```python
for school in schools:
    with open(sys.path[0] + f'/../data/{school}/{school}_w_profiles_processed.json', 'r') as f:
        the_school = json.load(f)
    dedupe(the_school)
    with open(sys.path[0] + f'/../data/{school}/{school}_w_profiles_processed.json', 'w') as f:
        json.dump(the_school, f)
```

```python
# labelled
for school in schools[:4]:
    with open(sys.path[0] + f'/../data/{school}/{school}_w_profiles_labeled.json', 'r') as f:
        the_school = json.load(f)
    dedupe(the_school)
    with open(sys.path[0] + f'/../data/{school}/{school}_w_profiles_labeled.json', 'w') as f:
        json.dump(the_school, f)
```
