```python
import sys
import json
import pandas as pd
import re
```

```python
def get_department_lk(url):
    return re.match("^(https|http)://[a-zA-Z0-9.-]*/", url).group(0)[:-1]
```

```python
def to_full_url(p, department_lk):
    lk = p.get('img')
    if lk is not None:
        if lk[0] == '/' and lk[:2] != '//':
            lk = department_lk + lk
            p.update({'img': lk})
```

```python
def clean_url(in_path, out_path):
    with open(in_path) as f:
        school_dict = json.load(f)
    for department in school_dict:
        if department.get('profiles') is not None:
            department_lk = get_department_lk(department.get('url'))
            for p in department.get('profiles'):
                to_full_url(p, department_lk)
    with open(out_path, 'w') as f:
        json.dump(school_dict, f)
    return school_dict
```

# Columbia

```python
with open(sys.path[0] + '/../data/columbia/columbia_test_sans_processing.json') as f:
    columbia = json.load(f)
```

```python
for department in columbia:
    if department.get('profiles') is not None:
        department_lk = get_department_lk(department.get('url'))
        for p in department.get('profiles'):
            to_full_url(p, department_lk)
```

```python
with open(sys.path[0] + '/../data/columbia_test_sans_processing_url', 'w') as f:
    json.dump(columbia, f)
```

# Dartmouth

```python
in_path = sys.path[0] + '/../data/dartmouth/dartmouth_test_sans_processing.json'
out_path = sys.path[0] + '/../data/dartmouth/dartmouth_test_sans_processing_url.json'
clean_url(in_path, out_path)
```
