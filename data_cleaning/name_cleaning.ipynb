```python
import sys
import json
import re
```

```python
with open(sys.path[0] + '/../data/columbia/columbia_test_sans_processing.json') as f:
    columbia = json.load(f)
```

```python
def process_name(p):
    name = p.get('name')
    first_name = None
    middle_name = None
    last_name = None
    if name is not None:
        name = re.sub(', Ph.D.', '', name)
        name = name.strip()
        if ',' in name:
            tmp = name.split(', ', 2)
            name = ' '.join([tmp[1], tmp[0]])
        names = name.split(' ')
        first_name = names[0]
        if len(names) == 2:
            middle_name = None
            last_name = names[1]
        elif len(names) == 3:
            middle_name = names[1]
            last_name = names[2]
    p.update({'name':name, 'first_name':first_name, 'middle_name': middle_name, 'last_name':last_name})
```

```python
for department in columbia:
    if department.get('profiles') is not None:
        for p in department.get('profiles'):
            process_name(p)
```
